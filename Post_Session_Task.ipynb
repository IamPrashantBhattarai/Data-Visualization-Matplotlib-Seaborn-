```python
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Check structure
tips.info()
```

```python
# Summary statistics
tips.describe()
```

```python
import matplotlib.pyplot as plt

# Distribution of a numeric variable
sns.histplot(tips['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# Count of a categorical variable
sns.countplot(x='day', data=tips)
plt.title("Count of Entries by Day")
plt.show()
```

```python
# Scatterplot for numeric variables
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Boxplot: numeric vs categorical
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Total Bill by Day")
plt.show()
```

```python
# Scatterplot with hue
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Tip vs Total Bill by Gender")
plt.show()
```

```python
# Pairplot for multiple numerical columns
sns.pairplot(tips, hue='sex')
plt.show()
```

Key insights form the above visualization:

Customers who spend more on the total bill tend to give higher tips.

Saturday has the highest number of entries, indicating it may be the busiest day.

Female customers tip slightly less on average than male customers.
